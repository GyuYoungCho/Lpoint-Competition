```
a = [1,2,3,4]
take = 3
def swap(num_list, a,b):
  if (a!=b):
    temp = num_list[a]
    num_list[a] = num_list[b]
    num_list[b] = temp

def permutate(num_list,take,k=0):
  if take==k:
    p = ""
    for j in range(take):
      p+=str(num_list[j])
    print(p);
    return 0;
  else:
    for i in range(k,len(num_list)):
      swap(num_list, k,i);
      permutate(num_list,take,k+1)
      swap(num_list, k,i);
permutate([1,2,3,4],3)
```
