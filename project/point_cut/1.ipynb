```python
f = open("transaction_info.csv", 'r', encoding='utf-8')

header = f.readline()

records = f.readlines()

groupByTransaction = dict()

for record in records:
    _, trans_id, _, _, pd_c, _, _, _, _ = record.strip().split(',')
    if pd_c != "unknown":
        trans_id, pd_c = eval(trans_id), eval(pd_c)
        groupByTransaction[trans_id] = groupByTransaction.get(trans_id, set())
        groupByTransaction[trans_id].add(pd_c)

relatedItems = dict()

for _, value in groupByTransaction.items():
    for product in value:
        relatedItems[product] = relatedItems.get(product, set())
        relatedItems[product] = relatedItems[product].union(value - set([product]))

# relatedItems 에 제품 : 함께 구매한 제품 으로 구성했습니다!

while True:
    product = input("구매상품: ")
    if product in relatedItems.keys():
        print(f"함께 구매한 상품: {relatedItems[product]}")
    else:
        print("해당 상품이 존재하지 않습니다.")      
        
relatedItems
```
