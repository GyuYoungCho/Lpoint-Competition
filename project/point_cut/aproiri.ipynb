```python
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import apriori
TFI = pd.read_csv("transfer.csv")
tfi_list = np.unique(TFI["trans_id"])
TFI.sort_values(by=["clnt_id","trans_id"])
```

```python
tfi1 = TFI[["trans_id","pd_c"]]
buy_list = []
for i in tfi_list:
    tfis = tfi1[tfi1["trans_id"]==i]
    if tfis.shape[0]==1:
        continue
    a= []
    for j in tfis["pd_c"]:
        a.append(j)
    a = np.unique(a)
    buy_list.append(a)
```

```python
buy_list
```

```python
from mlxtend.preprocessing import TransactionEncoder
te = TransactionEncoder()
te_ary = te.fit(buy_list).transform(buy_list)
df = pd.DataFrame(te_ary, columns=te.columns_)
```

```python
from mlxtend.frequent_patterns import association_rules
frequent_itemsets = apriori(df, min_support=0.01, use_colnames=True)
frequent_itemsets.sort_values(by="support")
```

```python
association_rules(frequent_itemsets, metric="confidence", min_threshold=0.2).sort_values(by="lift",ascending=False)
```

```python
OAI = pd.read_csv("Online.csv")
oai0 = OAI[OAI["action_type"]==0].reset_index(drop=True)
oai0[["clnt_id","sess_id"]] = oai0[["clnt_id","sess_id"]].astype(str)
oai0["oai_list"] = oai0["clnt_id"]+"_"+ oai0["sess_id"]
oai0.shape
```

```python
oai_list = np.unique(oai0["oai_list"])
oai1 = oai0[["oai_list","sech_kwd"]]
sech_list = []
for i in oai_list:
    oais = oai1[oai1["oai_list"]==i]
    if oais.shape[0]==1:
        continue
    a= []
    for j in oais["sech_kwd"]:
        a.append(j)
    a = np.unique(a)
    sech_list.append(a)
```

```python
oai2 = oai1["sech_kwd"].value_counts()
oai2=oai2[oai2 < 3]
for i in oai2.index:
    oai1.drop(oai1[oai1["sech_kwd"]==i].index)
oai1.shape
```

```python
sech_list
```

```python
te1 = TransactionEncoder()
te_ary1 = te1.fit(sech_list).transform(sech_list)
df1 = pd.DataFrame(te_ary1, columns=te.columns_)
```

```python
frequent_sechsets = apriori(df1, min_support=0.01, use_colnames=True)
frequent_sechsets.sort_values(by="support")
```
