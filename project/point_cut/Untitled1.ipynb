```python
import pandas as pd
import numpy as np
from tqdm import tqdm
```

```python
OAI = pd.read_csv("OnlineActionInfo.csv")
TFI = pd.read_csv("TransferInfo.csv")
OAI.head()
```

```python
print(len(TFI))
```

```python
oai2 = OAI[(OAI["action_type"]==0)|(OAI["action_type"]==6)][["action_type","clnt_id","sess_id", "sess_dt","hit_tm","trans_id","sech_kwd"]]
oai6 = OAI[OAI["action_type"]==6][["clnt_id","sess_id", "sess_dt","hit_tm","trans_id"]]
oai0 = OAI[OAI["action_type"]==0][["clnt_id","sess_id", "sess_dt","hit_tm","sech_kwd"]]
oai6 = oai6[oai6["trans_id"].notnull()]
```

```python
merge1 = TFI.merge(oai6,on = ["clnt_id","trans_id"])[["clnt_id","trans_id","pd_c", "sess_id", "sess_dt","hit_tm"]]
merge1 = TFI.merge(oai6, on=["clnt_id","trans_id"])[["clnt_id","trans_id","pd_c", "sess_id", "sess_dt","hit_tm"]]
merge1 = merge1[merge1["pd_c"]!="unknown"]
merge1 = merge1.drop_duplicates()
merge1.shape
```

```python
pd_c_list = {}

for row in tqdm(range(oai6.shape[0])):
    data = oai6.iloc[row,]
    sech = oai0[(oai0["clnt_id"]==data["clnt_id"])&(oai0["sess_id"]==data["sess_id"])]["sech_kwd"].reset_index(drop=True)
    for i in range(len(sech)):
        pd_c_list[data["trans_id"]] = pd_c_list.get(data["trans_id"], [])
        pd_c_list[data["trans_id"]].append(sech.iloc[i,])
```

```python
len(pd_c_list)
```

```python
tfi1 = TFI[TFI["pd_c"]!="unknown"]
pd_c_sech = {}
for trans, value in tqdm(pd_c_list.items()):
    tfi2 = tfi1[tfi1["trans_id"]==trans]["pd_c"].reset_index(drop=True)
    for i in tfi2:
        for kwd in value:
            pd_c_sech[i] = pd_c_sech.get(i, [])
            pd_c_sech[i].append(kwd)
```

```python
from collections import Counter
related_pd_c = {}
for pdc, value in tqdm(pd_c_sech.items()):
    counts = Counter(value)
    max_count = 0
    for co in counts:
        if counts[co] > max_count:
            max_count = counts[co]
            max_item = co
    related_pd_c[pdc] = related_pd_c.get(pdc, [])
    related_pd_c[pdc].append(max_item)
```
